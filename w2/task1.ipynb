```python
import networkx as nx
import numpy as np
import pandas as pd
from tqdm import tqdm_notebook
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
G = nx.Graph([(0,1), (1,2), (2,0)])

deg_2_all = [3]
share_deg_2_all = [1]
for index, i in tqdm_notebook(enumerate(range(2, 3002, 2))):
    number_node = index + 3
    len_edges = len(list(G.edges()))
    rnd_int = np.random.random_integers(0, len_edges-1, 1)[0]
    edge = list(G.edges())[rnd_int]
    G.add_node(number_node)
    G.add_edges_from([(number_node, edge[0]), (number_node, edge[1])])
    
    deg_2 = 0
    for j in range(len(G.nodes())):
        if G.degree[j] == 2:
            deg_2 += 1
            
    share_deg_2_all.append(deg_2 / len(G.nodes))
    deg_2_all.append(deg_2)
    
            


    
```

```python
plt.figure(figsize=(10, 5))
plt.plot(deg_2_all, color='r')
plt.ylabel('number of nodes of degree two')
plt.xlabel('nodes')
plt.show()
```

```python
plt.figure(figsize=(15, 5))
plt.plot(share_deg_2_all, color='r')
plt.ylabel('share of nodes of degree two')
plt.xlabel('nodes')
plt.show()
```
